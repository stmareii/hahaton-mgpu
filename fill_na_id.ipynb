## Заполняем пропущенные id_doc 

1. Мы выяснили, что есть N/a, и есть "-", превратим их в n/a

```python
df_id_full = df.copy()
df_id_full["id_doc"] = df_id_full["id_doc"].replace("-", np.nan)
```

```python
counts = df_id_full["id_doc"].value_counts()
counts
```

```python
df_id_full.isna().sum()
```

2. Теперь разделим датасет на два датафрейма:
- один будет с id_doc == na
- второй будет просто без них

```python
df_with_na = df_id_full[df_id_full["id_doc"].isna()]
df_with_na = df_with_na.reset_index(drop=True)

df_without_na = df_id_full[df_id_full["id_doc"].notna()]
df_without_na = df_without_na.reset_index(drop=True)
```
